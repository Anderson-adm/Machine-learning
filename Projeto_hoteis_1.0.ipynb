# 1 - **Definição do problema**

### 1.1 - **Apresentando o conjunto de dados**

Este conjunto de dados contém informações de reserva de um hotel urbano e de um hotel resort e inclui informações como quando a reserva foi feita, duração da estadia, número de adultos, crianças e / ou bebês e número de vagas disponíveis, entre outras coisas.

Todas as informações de identificação pessoal foram removidas dos dados.

 ### 1.2 - **O que cada variável representa**

- **hotel:**  H1 = Hotel resort ou H2 = Hotel urbano

- **esta cancelado:** Valor indicando se a reserva foi cancelada (1) ou não (0)

- **tempo de espera:** Número de dias decorridos entre a data de entrada da reserva no PMS e a data de chegada

- **arrival_date_year:** Data de chegada

- **arrival_date_month:** Mês da data de chegada

- **arrival_date_week_number:** Número da semana do ano para a data de chegada

- **arrival_date_day_of_month:** Data de chegada

- **nights_in_weekend_nights:** 
> Número de noites de fim de semana (sábado ou domingo) em que o hóspede ficou ou reservou para ficar no hotel

**nights_in_week_nights:** 
> Número de noites da semana (de segunda a sexta-feira) em que o hóspede ficou ou reservou para ficar no hotel

- **adultos:** Número de Adultos

- **crianças:** Número de filhos

- **bebês:** Número de bebês

- **refeição:** 
    Tipo de refeição reservada. As categorias são apresentadas em pacotes de refeições de hospitalidade padrão: 
   > * Indefinido / SC - sem pacote de refeições; 
   > * BB - Alojamento e café da manhã; 
   > * HB - Meia pensão (café da manhã e outra refeição - geralmente jantar); 
   > * FB - Pensão completa (café da manhã, almoço e jantar)

- **país:** País de origem. As categorias são representadas no formato ISO 3155–3: 2013

- **segmento de mercado:** 
> Designação do segmento de mercado. Nas categorias, o termo "TA" significa "agentes de viagens" e "TO" significa "operadores turísticos"

- **canal de distribuição:** 
> Canal de distribuição de reservas. O termo "TA" significa "agentes de viagens" e "TO" significa "operadores turísticos"

- **is_repeated_guest:** Valor indicando se o nome da reserva era de um hóspede repetido (1) ou não (0)

- **cancelamentos anteriores:** Número de reservas anteriores que foram canceladas pelo cliente antes da reserva atual

- **previous_bookings_not_canceled:**  Número de reservas anteriores não canceladas pelo cliente antes da reserva atual

- **Reserved_room_type:**  Código do tipo de quarto reservado. O código é apresentado em vez da designação por razões de anonimato.

- **assign_room_type:** 
> Código para o tipo de quarto atribuído à reserva. Às vezes, o tipo de quarto atribuído difere do tipo de quarto reservado devido a razões de operação do hotel (por exemplo, reservas em excesso) ou por solicitação do cliente. O código é apresentado em vez da designação por razões de anonimato.

- **booking_changes:** 
> Número de alterações / alterações feitas na reserva desde o momento em que a reserva foi inserida no PMS até o momento do check-in ou cancelamento

- **deposit_type:** 
Indicação sobre se o cliente fez um depósito para garantir a reserva. Essa variável pode assumir três categorias: 
> * Nenhum depósito - nenhum depósito foi feito; 
> * Não reembolso - foi feito um depósito no valor do custo total da estadia; 
> * Reembolsável - foi feito um depósito com um valor abaixo do custo total da estadia.

- **agente:** ID da agência de viagens que fez a reserva

- **companhia:** ID da empresa / entidade que fez a reserva ou responsável pelo pagamento da reserva. O ID é apresentado em vez da designação por razões de anonimato

- **days_in_waiting_list** Número de dias em que a reserva estava na lista de espera antes de ser confirmada ao cliente

- **Tipo de Cliente** Tipo de reserva, assumindo uma das quatro categorias:
> * Contrato - quando a reserva tem um lote ou outro tipo de contrato associado a ela; 
> * Grupo - quando a reserva está associada a um grupo; 
> * Transitório - quando a reserva não faz parte de um grupo ou contrato e não está associado a outra reserva transitória; 
> * Transient-party - quando a reserva é transitória, mas está associada a pelo menos outra reserva transitória

- **adr:** 
>Taxa diária média, conforme definida pela divisão da soma de todas as transações de hospedagem pelo número total de noites de permanência

- **required_car_parking_spaces:** Número de vagas de estacionamento exigidas pelo cliente

- **total_of_special_requests:** Número de pedidos especiais feitos pelo cliente (por exemplo, cama de solteiro ou andar alto)

- **reservation_status:** Último status da reserva, assumindo uma de três categorias: 
> * Cancelado - a reserva foi cancelada pelo cliente; 
> * Check-Out - o cliente fez o check-in, mas já partiu; 
> * No-Show - o cliente não fez o check-in e informou o hotel do motivo

- **reservation_status_date:** Data em que o último status foi definido.

# **Pré-processamento**

```python
hoteis.head()
```

```python
hoteis['hotel'].replace({'Resort Hotel': 0, 'City Hotel': 1}, inplace=True)
```

```python
tipos = hoteis.dtypes
objeto = tipos[tipos != 'object']

lista_n_objetos = list(objeto.index)
```

```python
lista_n_objetos
```

```python
droped = hoteis.drop(['hotel', 'is_canceled', 'lead_time', 'arrival_date_year', 'arrival_date_week_number', 'arrival_date_day_of_month',
                     'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children', 'babies', 'is_repeated_guest', 'previous_cancellations',
                     'previous_bookings_not_canceled', 'booking_changes', 'agent', 'days_in_waiting_list', 'adr', 'required_car_parking_spaces',
                     'total_of_special_requests', 'reservation_status', 'reservation_status_date'], axis=1)
```

```python
hoteis.head()
```

```python
droped.head()
```

```python
concat = pd.get_dummies(droped)
```

```python
concat.head()
```

```python
reservas = pd.concat([hoteis, concat], axis=1)
```

```python
reservas.head()
```

```python
reservas.drop(['arrival_date_month', 'meal', 'country', 'market_segment', 'distribution_channel', 'reserved_room_type', 'distribution_channel',
               'reserved_room_type', 'assigned_room_type', 'deposit_type', 'customer_type'], axis=1, inplace=True)
```

```python
reservas.drop(['reservation_status', 'reservation_status_date'], axis=1, inplace=True)
```

```python
reservas.head()
```

```python
y = reservas['hotel']
x = reservas.drop('hotel', axis=1)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler

normalizador = MinMaxScaler(feature_range=(0, 1))
x_norm = normalizador.fit_transform(x)

pca = PCA(n_components= 60)
x_pca = pca.fit_transform(x_norm)
```

```python
print('Variância explicada dos componentes: ', sum(pca.explained_variance_ratio_))
```

```python
x_pca
```

```python
# Importando funções de classificação:
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC

# Importando validação cruzada:
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

# criando modelos:

logistic = LogisticRegression(n_jobs=-1)
gaussian = GaussianNB()
tree = DecisionTreeClassifier()
knn = KNeighborsClassifier(n_jobs=-1)
forest = RandomForestClassifier(n_jobs=-1)
extra = ExtraTreesClassifier(n_jobs=-1)
ada = AdaBoostClassifier()
gradient = GradientBoostingClassifier()
bagging = BaggingClassifier(n_jobs=-1)
vector = SVC()

skfold = StratifiedKFold(n_splits=3)

# Criando validação cruzada:
logistic_cv = cross_val_score(logistic, x_pca, y, cv=skfold, n_jobs=-1)
gaussian_cv = cross_val_score(gaussian, x_pca, y, cv=skfold, n_jobs=-1)
tree_cv = cross_val_score(tree, x_pca, y, cv=skfold, n_jobs=-1)
knn_cv = cross_val_score(knn, x_pca, y, cv=skfold, n_jobs=-1)
forest_cv = cross_val_score(forest, x_pca, y, cv=skfold, n_jobs=-1)
extra_cv = cross_val_score(extra, x_pca, y, cv=skfold, n_jobs=-1)
ada_cv = cross_val_score(ada, x_pca, y, cv=skfold, n_jobs=-1)
gradient_cv = cross_val_score(gradient, x_pca, y, cv=skfold, n_jobs=-1)
bagging_cv = cross_val_score(bagging, x_pca, y, cv=skfold, n_jobs=-1)
vector_cv = cross_val_score(vector, x_pca, y, cv=skfold, n_jobs=-1)

# Exibindo resultados:
print('logistic: ', logistic_cv.mean())
print('gaussian: ', gaussian_cv.mean())
print('tree: ', tree_cv.mean())
print('knn: ', knn_cv.mean())
print('forest: ', forest_cv.mean())
print('extra: ', extra_cv.mean())
print('ada: ', ada_cv.mean())
print('gradient: ', gradient_cv.mean())
print('bagging: ', bagging_cv.mean())
print('vector: ', vector_cv.mean())
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold
import numpy as np

# Criando parâmetros a serem variados no grid:
c = np.array([1.0, 0.95, 1.05, 1.1, 1.2, 2.0])
kernel = ['linear', 'poly', 'rbf', 'sigmoid']
polinomio = np.array([2, 3, 4])
gama = ['scale', 'auto']
valores_grid = {'C': c, 'kernel': kernel, 'degree': polinomio, 'gamma': gama}

# Criando modelo:
modelo = SVC()

# Criando validação cruzada:
skfold = StratifiedKFold(n_splits=3, shuffle=True)

svc_grid = GridSearchCV(estimator=modelo, param_grid=valores_grid, cv=skfold, n_jobs=-1)
svc_grid.fit(x_pca, y)

# Imprimindo resultados:
print('Melhor valor constante: ', svc_grid.best_estimator_.C)
print('Melhor Kernel: ', svc_grid.best_estimator_.kernel)
print('Melhor grau polinomio: ', svc_grid.best_estimator_.degree)
print('Melhor gama: ', svc_grid.best_estimator_.gamma)
print('Melhor score: ', svc_grid.best_score_)
```
