## 1 - **Definição do problema**

### 1.1 - **Apresentando o conjunto de dados**

Este conjunto de dados contém informações de reserva de um hotel urbano e de um hotel resort e inclui informações como quando a reserva foi feita, número de adultos, crianças  e número de vagas disponíveis, entre outras coisas.

Todas as informações de identificação pessoal foram removidas dos dados.

### 1.2 - **O que cada variável representa**

- **hotel:**  
H1 = Hotel resort ou H2 = Hotel urbano

- **is_canceled:**  
Valor indicando se a reserva foi cancelada (1) ou não (0)

- **lead_time:**  
Número de dias decorridos entre a data de entrada da reserva no PMS e a data de chegada

- **arrival_date_year:**  
Data de chegada

- **arrival_date_month:**  
Mês da data de chegada

- **arrival_date_week_number:**  
Número da semana do ano para a data de chegada

- **arrival_date_day_of_month:**  
Data de chegada

- **nights_in_weekend_nights:**  
 Número de noites de fim de semana (sábado ou domingo) em que o hóspede ficou ou reservou para ficar no hotel

- **nights_in_week_nights:**  
Número de noites da semana (de segunda a sexta-feira) em que o hóspede ficou ou reservou para ficar no hotel

- **adults:**  
Número de Adultos

- **children:**  
Número de filhos

- **babies:**  
Número de bebês

- **meal:**  
Tipo de refeição reservada. As categorias são apresentadas em pacotes de refeições de hospitalidade padrão: 
   > * Indefinido / SC - sem pacote de refeições; 
   > * BB - Alojamento e café da manhã; 
   > * HB - Meia pensão (café da manhã e outra refeição - geralmente jantar); 
   > * FB - Pensão completa (café da manhã, almoço e jantar)

- **country:**  
País de origem. As categorias são representadas no formato ISO 3155–3: 2013

- **market_segment:**  
Designação do segmento de mercado. Nas categorias, o termo "TA" significa "agentes de viagens" e "TO" significa "operadores turísticos"

- **distribution_channel:**  
Canal de distribuição de reservas. O termo "TA" significa "agentes de viagens" e "TO" significa "operadores turísticos"

- **is_repeated_guest:**  
Valor indicando se o nome da reserva era de um hóspede repetido (1) ou não (0)

- **previous_cancellations:**  
Número de reservas anteriores que foram canceladas pelo cliente antes da reserva atual

- **previous_bookings_not_canceled:**  
Número de reservas anteriores não canceladas pelo cliente antes da reserva atual

- **Reserved_room_type:**  
Código do tipo de quarto reservado. O código é apresentado em vez da designação por razões de anonimato.

- **assign_room_type:** 
> Código para o tipo de quarto atribuído à reserva. Às vezes, o tipo de quarto atribuído difere do tipo de quarto reservado devido a razões de operação do hotel (por exemplo, reservas em excesso) ou por solicitação do cliente. O código é apresentado em vez da designação por razões de anonimato.

- **booking_changes:** 
> Número de alterações / alterações feitas na reserva desde o momento em que a reserva foi inserida no PMS até o momento do check-in ou cancelamento

- **deposit_type:**  
Indicação sobre se o cliente fez um depósito para garantir a reserva. Essa variável pode assumir três categorias: 
> * Nenhum depósito - nenhum depósito foi feito; 
> * Não reembolso - foi feito um depósito no valor do custo total da estadia; 
> * Reembolsável - foi feito um depósito com um valor abaixo do custo total da estadia.

- **agent:**  
ID da agência de viagens que fez a reserva

- **company:**  
ID da empresa / entidade que fez a reserva ou responsável pelo pagamento da reserva. O ID é apresentado em vez da designação por razões de anonimato

- **days_in_waiting_list**  
Número de dias em que a reserva estava na lista de espera antes de ser confirmada ao cliente

- **Tipo de Cliente**  
Tipo de reserva, assumindo uma das quatro categorias:
> * Contrato - quando a reserva tem um lote ou outro tipo de contrato associado a ela; 
> * Grupo - quando a reserva está associada a um grupo; 
> * Transitório - quando a reserva não faz parte de um grupo ou contrato e não está associado a outra reserva transitória; 
> * Transient-party - quando a reserva é transitória, mas está associada a pelo menos outra reserva transitória

- **adr:** 
>Taxa diária média, conforme definida pela divisão da soma de todas as transações de hospedagem pelo número total de noites de permanência

- **required_car_parking_spaces:**  
Número de vagas de estacionamento exigidas pelo cliente

- **total_of_special_requests:**  
Número de pedidos especiais feitos pelo cliente (por exemplo, cama de solteiro ou andar alto)

- **reservation_status:**  
Último status da reserva, assumindo uma de três categorias: 
> * Cancelado - a reserva foi cancelada pelo cliente; 
> * Check-Out - o cliente fez o check-in, mas já partiu; 
> * No-Show - o cliente não fez o check-in e informou o hotel do motivo

- **reservation_status_date:**  
Data em que o último status foi definido.

```python
# Investigando dados faltantes nos dois datasets:
def faltantes_percentual(data):
    """ 
    Uma função que retorna as colunas de um dataset com mais de 0%
    faltantes.
    """

    faltando = data.isnull().sum()
    percentual = (faltando / len(data.iloc[:, 0])) * 100
    maior_zero = percentual[percentual > 0]
    return maior_zero
```

### 2.2 - **Perguntas Iniciais**

* **Qual o mes e o trimistre que possuem a maior quantidade de cancelamentos?**

```python
# Criando função
def cancelamentos(data):
    """ Uma função que retorna a % de cancelamentos por mes."""

    agrupado = data.loc[:, ['is_canceled', 'arrival_date_month']].copy()
    cancelados = agrupado.loc[agrupado['is_canceled'] == 1].copy()
    cancelado_mensal = cancelados['arrival_date_month'].value_counts(normalize=True) * 100
    return cancelado_mensal

# Chamando a função criada
cancelado_urbano = cancelamentos(urbano)
cancelado_resort = cancelamentos(resort)

# Plotando o resultado
meses_ordenados = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

plt.figure(figsize=(25, 8))
plt.subplot(2, 1, 1)
sns.barplot(x = cancelado_urbano.index, y = cancelado_urbano.values, palette='Blues_d', order=meses_ordenados)
plt.title('%  cancelamentos por mes - Urbano e Resort',fontsize=16)
#plt.gca().xaxis.set_major_locator(plt.NullLocator())

plt.subplot(2, 1, 2)
sns.barplot(x = cancelado_resort.index, y = cancelado_resort.values, palette = 'Blues_d', order= meses_ordenados)
plt.show()
```

* **No Urbano não há nenhuma sazonalidade, mas os cancelamentos parecem aumentar  ao longo do ano e ficando maiores a partir de abril até setembro, onde começa a cair até dezembro.**

* **No resort, o percentual de cancelamento por mes é mais semelhante entre os meses, exceto porjulho e agosto que se destacam dos demais.**

* **É possível notar diferenças entre a quantidade de cancelamentos por mes em cada hotel, porém, talvez seja mais esclarecedor olhar essa mesma informação no formato de trimestre.**


```python
def cancelamentos_trimestre(data):
    """ ."""

    agrupado = data.loc[:, ['is_canceled', 'arrival_date_month']].copy()
    
    cancelados = agrupado.loc[agrupado['is_canceled'] == 1].copy()
    
    cancelados['id_mes'] = cancelados.loc[:, 'arrival_date_month'].map({'January': 1, 'February': 2, 'March': 3, 'April': 4, 
                                                                           'May': 5, 'June': 6, 'July': 7, 'August': 8, 
                                                                          'September': 9, 'October': 10, 'November': 11, 'December': 12})
    
    cancelados['trimestre'] = data.loc[:, 'arrival_date_month'].map({'January': 'primeiro', 'February': 'primeiro', 'March': 'primeiro',
                                                                'April': 'segundo', 'May': 'segundo', 'June': 'segundo',
                                                               'July': 'terceiro', 'August': 'terceiro', 'September': 'terceiro',
                                                               'October': 'quarto', 'November': 'quarto', 'December': 'quarto'})
   
    cancelados.sort_values(by='id_mes', inplace=True)
    
    cancelado_trimestre = cancelados['trimestre'].value_counts(normalize=True) * 100
    
    return cancelado_trimestre


# chamando função
cancelado_urbano_trimestre = cancelamentos_trimestre(urbano)
cancelado_resort_trimestre = cancelamentos_trimestre(resort)

# Visualizando resultado
ordem = ['primeiro', 'segundo', 'terceiro']
sns.set()
plt.figure(figsize=(25, 8))
plt.subplot(1, 2, 1)
sns.barplot(x= cancelado_urbano_trimestre.index, y= cancelado_urbano_trimestre.values, order= ordem, palette='Blues')
plt.title('%  cancelamentos por trimestre - Urbano',fontsize=16)

plt.subplot(1, 2, 2)
sns.barplot(x= cancelado_resort_trimestre.index, y= cancelado_resort_trimestre.values, order= ordem, palette='Blues');
plt.title('% cancelamentos por trimestre - Resort',fontsize=16)
plt.show()
```

* **Agora parece ter ficado mais claro que no hotel urbano os cancelamentos se concentram no segundo e terceiro semestre, com destaque para o segundo reponsável por  mais de 30% dos cancelamentos anuais**
* **Enquanto que o resort espelha esse comportamento, porém com uma concentração muito maior no terceiro, concentrando 35% dos cancelamentos anuais.**

* **Qual o número de reservas por ano que são feitas com antecedência de: 15, 30, 60, 90, 120, 150, 180 e + de 6 meses**

```python
def bins(data1, data2):
    """ 
    Recebe os datasets e divide a variável tempo de esperar em intervalos de dias param serem analisados.
    """

    intervalos_1 = pd.cut(data1['lead_time'], [0, 15, 30, 45, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330, 360])
    intervalos_2 = pd.cut(data2['lead_time'], [0, 15, 30, 45, 60, 90, 120, 150, 180, 210, 240, 270, 300, 330, 360])
    
    percent_1 = intervalos_1.value_counts(normalize=True) * 100
    percent_2 = intervalos_2.value_counts(normalize=True) * 100  


    plt.figure(figsize=(25, 12))
    plt.subplot(1, 2, 1)
    sns.barplot(y=percent_1.index, x=percent_1.values, palette='Blues_d')
    plt.title('% reservas por intervalo - Urbano', fontsize=16)
    
    plt.subplot(1, 2, 2)
    sns.barplot(y=percent_2.index, x=percent_2.values, palette='Blues_d')
    plt.title('% reservas por intervalo - Resort', fontsize=16)
    plt.show()
```

```python
bins(urbano, resort)
```

* Percebemos que o dois hotéis se assemelham e a maioria dos seus clientes compram a reserva com até 15 dias de antecedência.
* Porém, para o urbano o segundo maior grupo é daqueles que compram entre 2 e 3 meses de antecedência.

* Usaremos o método do cotovelo, que é a análise gráfica do número de clusters que tenha a menor soma dos quadrados intra-clusters.
* Em outras palavras, estamos buscando uma quantidade de clusters que a partir dali aumentar o número de clusters não faz mais os centróides se moverem de forma significativa
* Esse ponto pode ser o ideal para utilizarmos na divisão de nossos grupos

##### **Clustering urbano**

```python
# TESTANDO 100 VALORES DIFERENTES PARA A QUANTIDADE DE CLUSTERINGS PARA TENTAR ESTIMAR O NÚMERO IDEAL:

from sklearn.cluster import KMeans

sse_urbano = [] # lista que armazenará a soma dos quadadros da distância dos pontos aos centróides em cada cluster
list_k = list(range(1, 20)) # intervalo de clusters a serem criados

for k in list_k:
    km = KMeans(n_clusters=k, n_jobs=-1)
    km.fit(urbano_norm)
    sse_urbano.append(km.inertia_)

# EXIBINDO OS RESULTADOS:
plt.figure(figsize=(10, 9))
plt.plot(list_k, sse_urbano, '-o')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of squared distance');
```

* Parece que um valor interessante para o dataset urbano é cerca de 20 clusters
* Apenas observando o gráfico podem surgir dúvidas quanto qual ponto é realmente o ideal, para auxiliar existe uma fórmula que calcula a distância de todos os pontos da curva à uma reta que conecta o primeiro e o último ponto
* Nesse cálculo, estamos interessados no maior resultado possível, esse significa o nosso ponto ideal

```python
# EXIBINDO A RETA QUE CONECTA O PRIMEIRO E ÚLTIMO PONTO DA CURVA:

x_coordenadas = [list_k[0], list_k[-1]]
y_coordenadas = [sse_urbano[0], sse_urbano[-1]]

plt.figure(figsize=(10, 9))
plt.plot(list_k, sse_urbano, '-o')
plt.plot(x_coordenadas, y_coordenadas, '--r')
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of squared distance');
```

* A fórmula que calcula a distancia de um ponto a essa reta é a seguinte: 

```python
# CRIANDO FUNÇÃO PARA A FÓRMULA:

def melhor_n_clusters(sse_urbano):
    x1, y1 = 1, sse_urbano[0]
    x2, y2 = 30, sse_urbano[len(sse_urbano)-1]

    distancias = []
    for i in range(len(sse_urbano)):
        x0 = i+1
        y0 = sse_urbano[i]
        numerator = abs((y2-y1)*x0 - (x2-x1)*y0 + x2*y1 - y2*x1)
        denominator = np.sqrt((y2 - y1)**2 + (x2 - x1)**2)
        distancias.append(numerator/denominator)
    
    return distancias.index(max(distancias)) + 1
```

```python
melhor_n_clusters(sse_urbano)
```

* Portanto, segundo o método utilizado o número ideal de clusters para o dataset urbano é 24

```python
# IMPORTANDO FUNÇÕES DE CLASSIFICAÇÃO:

from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.svm import SVC

# Importando validação cruzada:

from sklearn.model_selection import cross_val_score
from sklearn.model_selection import StratifiedKFold

# CRIANDO MODELOS:

logistic = LogisticRegression(n_jobs=-1)
gaussian = GaussianNB()
tree = DecisionTreeClassifier()
knn = KNeighborsClassifier(n_jobs=-1)
forest = RandomForestClassifier(n_jobs=-1)
extra = ExtraTreesClassifier(n_jobs=-1)
ada = AdaBoostClassifier()
gradient = GradientBoostingClassifier()
bagging = BaggingClassifier(n_jobs=-1)
vector = SVC()

skfold = StratifiedKFold(n_splits=3, shuffle=True)

# CRIANDO VALIDAÇÃO CRUZADA:

logistic_cv = cross_val_score(logistic, x_pca, y, cv=skfold, n_jobs=-1)
gaussian_cv = cross_val_score(gaussian, x_pca, y, cv=skfold, n_jobs=-1)
tree_cv = cross_val_score(tree, x_pca, y, cv=skfold, n_jobs=-1)
knn_cv = cross_val_score(knn, x_pca, y, cv=skfold, n_jobs=-1)
forest_cv = cross_val_score(forest, x_pca, y, cv=skfold, n_jobs=-1)
extra_cv = cross_val_score(extra, x_pca, y, cv=skfold, n_jobs=-1)
ada_cv = cross_val_score(ada, x_pca, y, cv=skfold, n_jobs=-1)
gradient_cv = cross_val_score(gradient, x_pca, y, cv=skfold, n_jobs=-1)
bagging_cv = cross_val_score(bagging, x_pca, y, cv=skfold, n_jobs=-1)
vector_cv = cross_val_score(vector, x_pca, y, cv=skfold, n_jobs=-1)

# EXIBINDO RESULTADOS:

print('logistic: ', logistic_cv.mean())
print('gaussian: ', gaussian_cv.mean())
print('tree: ', tree_cv.mean())
print('knn: ', knn_cv.mean())
print('forest: ', forest_cv.mean())
print('extra: ', extra_cv.mean())
print('ada: ', ada_cv.mean())
print('gradient: ', gradient_cv.mean())
print('bagging: ', bagging_cv.mean())
print('vector: ', vector_cv.mean())
```
